```python
from sklearn.svm import SVC, SVR
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn
```

```python
iris = datasets.load_iris(as_frame = True).frame
```

```python
iris.head()
```

```python
iris.isnull().sum()
```

```python
X = iris.drop('target', axis = 1)
y = iris['target']
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.3, random_state = 42, stratify = y
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train, y_train)
X_test_scaled = scaler.transform(X_test)
```

```python
svc = SVC()
svc.fit(X_train_scaled, y_train)
```

```python
y_pred = svc.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, classification_report, f1_score

print("Accuracy : ", accuracy_score(y_test, y_pred))
print("Classification : ", classification_report(y_test, y_pred))
```

```python
svc = SVC(kernel = 'linear')
svc.fit(X_train_scaled, y_train)
y_pred = svc.predict(X_test_scaled)
print("Accuracy : ", accuracy_score(y_test, y_pred))
print("Classification : ", classification_report(y_test, y_pred))
```

```python
svc = SVC(kernel = 'rbf')
svc.fit(X_train_scaled, y_train)
y_pred = svc.predict(X_test_scaled)
print("Accuracy : ", accuracy_score(y_test, y_pred))
print("Classification : ", classification_report(y_test, y_pred))
```

```python
C_vals = [0.5,1,2,3,4,5]

for c in C_vals:
    svc = SVC(kernel = 'rbf', C = c)
    svc.fit(X_train_scaled, y_train)
    y_pred = svc.predict(X_test_scaled)
    print("C : ", c, "Accuracy : ", accuracy_score(y_test, y_pred))
    # print("Classification : ", classification_report(y_test, y_pred))
```

```python
data = datasets.load_diabetes(as_frame = True).frame
```

```python
data.head()
```

```python
X = data.drop('target', axis = 1)
y = data['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.3, random_state = 42
)
```

```python
y_scaler = StandardScaler()

y_train_scaled = y_scaler.fit_transform(y_train.values.reshape(-1, 1)).ravel()
y_test_scaled = y_scaler.transform(y_test.values.reshape(-1, 1)).ravel()
```

```python
model = SVR()

model.fit(X_train, y_train_scaled)
```

```python
y_pred_scaled = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print("r2 score : ", r2_score(y_test_scaled, y_pred_scaled))
```

```python
model = SVR(kernel = 'linear')

model.fit(X_train, y_train_scaled)
y_pred_scaled = model.predict(X_test)
print("r2 score : ", r2_score(y_test_scaled, y_pred_scaled))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "C"  :[ 1, 2, 5, 10, 50],
    "kernel" : ['rbf', 'linear'],
    'epsilon' : [0.01, 0.1, 0.2, 0.3, 0.5]
    }

grid = GridSearchCV(
    SVR(),
    param_grid,
    cv = 5,
    scoring = 'r2'
)

grid.fit(X_train, y_train_scaled)
```

```python
y_pred = grid.predict(X_test)
```

```python
print("r2 score : ", r2_score(y_test_scaled, y_pred))
```

```python
print(grid.best_params_)
```

```python
from sklearn.svm import LinearSVR

model = LinearSVR(C = 10, epsilon = 0.1, max_iter = 10000)
model.fit(X_train, y_train_scaled)
y_pred_scaled = model.predict(X_test)
print("r2 score : ", r2_score(y_test_scaled, y_pred_scaled))
```
